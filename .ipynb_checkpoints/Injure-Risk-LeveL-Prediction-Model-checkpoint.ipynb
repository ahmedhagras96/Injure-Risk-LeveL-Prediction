**Now we need to spilt data into train and test data to make sure our model is working will and get the accurcy of our model
and we will take 20% of data as test data**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=90)
```

```python
training_data = pd.concat([x_train, y_train], axis=1)
training_data
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

**In nxt few cells we want to get most accurcy score we can get so we assume K value in range 1 to 50 and get score in allthese cases and get max score by argmax and reapply KNN model with K value that make the max score**

```python
scores = []
for k in range(1, 50):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    score = accuracy_score(y_test, y_pred)
    scores.append(score)
```

```python
scores
```

```python
np.argmax(scores) + 1
```

```python
np.max(scores)
```

```python
plt.plot(range(1, 50), scores)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=np.argmax(scores) + 1)
```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```
